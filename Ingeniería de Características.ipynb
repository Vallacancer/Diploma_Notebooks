```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
fut = pd.DataFrame({"Posición":["LW","RW","LW","ST","GK"],
                    "Número":[7,10,11,9,1],
                    "Altura":[185,170,174,182,193],
                    "Peso":[80,72,68,85,92],
                    "Rating":[94,93,92,92,92]})

fut.corr()
```

```python
df = fut.groupby("Altura").sum()

df.reset_index()
```

```python
fut.corr()
```

```python
reg = linear_model.Ridge(alpha=5)
X=fut[["Número", "Altura", "Peso"]]
Y=fut["Rating"]

reg.fit(X, Y)
reg.score(X,Y)
```
