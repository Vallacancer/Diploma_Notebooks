```python
import numpy as np

listadatos = [45, 67, 23, 51, 89, 100, 23, 46, 57, 12, 1, 87, 23, 69, 33, 18, 87, 45, 73, 28]
MatrizA = np.array([[34, 61, 0, 12], [98, 53, 76, 2], [19, 56, 87, 32], [100, 62, 91, 21], [84, 50,41, 3]])

matrizB = np.array(listadatos).reshape(5,4)

print(MatrizA, "\n")
print(matrizB)
print("min: ", MatrizA.min())
print("max: ", MatrizA.max())
print("Promedio columnas: ", MatrizA.mean(axis=0))
print("Suma elementos segunda y tercera fila: ", MatrizA[1:3].sum())
print(MatrizA + matrizB, "\n")

print(MatrizA @ matrizB.reshape(4,5), "\n")

print(np.transpose(matrizB.reshape(4,5)))
```

```python
import pandas as pd
import numpy as np

def horario_programado(base, puerto, serie, compañia="Ultramar"):
  for indice_base in base.query('(Embarcadero == @puerto) & (Ruta == "IM-CUN") & (Proveedor == @compañia)')["Horario real"].index:
    chequeo = np.timedelta64(23,"h")
    for indice_serie in serie.index:
      delta = abs(base["Horario real"][indice_base] - np.timedelta64(30, "m") - serie[indice_serie])
      if delta < chequeo:
        chequeo = delta
        base.loc[indice_base, "Horario programado"] = serie[indice_serie]


#Importar base y convertir datos a timedelta
base = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DataSets/Indice/Octubre/Xounting_Base_Oct.xlsx", sheet_name="Hoja1",parse_dates=["Fecha de muestra","Horario real", "Horario programado"])
# base["Horario real"] = pd.to_timedelta(base["Horario real"].astype(str))
# base["Horario programado"] = pd.to_timedelta(base["Horario programado"].astype(str))

#Se crean los horarios y se vuelven fechas
playaTortugas = pd.to_datetime(pd.Series(["10:30:00", "11:45:00", "13:00:00", "14:30:00", "16:00:00", "17:30:00", "19:00:00"]))
playaCaracol = pd.to_datetime(pd.Series(["10:30:00","11:45:00","13:00:00","14:30:00","16:00:00","17:30:00","19:00:00"]))
PuertoJuarezUltramar = pd.to_datetime(pd.Series(["6:00:00","7:00:00","7:30:00","8:00:00","8:30:00","9:00:00","9:30:00","10:00:00",
                                  "10:30:00","11:00:00","11:30:00","12:00:00","12:30:00","13:00:00","13:30:00","14:00:00",
                                  "15:00:00","16:00:00","16:30:00","17:00:00","17:30:00","18:00:00","18:30:00","19:00:00",
                                  "19:30:00","20:00:00","21:00:00","22:30:00","00:00:00"]))
PuertoJuarezJetway = pd.to_datetime(pd.Series(["7:00:00", "8:00:00", "9:00:00", "9:45:00", "10:30:00", "11:15:00", "12:00:00",
                                               "13:00:00", "14:00:00", "15:00:00", "16:00:00", "17:00:00", "18:00:00", "19:00:00",
                                               "20:00:00", "21:00:00"]))

#Cambiar la serie a tiempo
#playaTortugas = pd.to_datetime(playaTortugas)

#Sumarle 1 hora
#playaTortugas = playaTortugas + np.timedelta64(1, "h")

#Accesar los datos de Playa Tortugas IM-CUN
#base.query('(Embarcadero == "Playa Tortugas") & (Ruta == "IM-CUN")')["Horario real"]

horario_programado(base, "Playa Tortugas", playaTortugas)
horario_programado(base, "Playa Caracol", playaCaracol)
horario_programado(base, "Puerto Juarez", PuertoJuarezUltramar)
horario_programado(base, "Puerto Juarez", PuertoJuarezJetway, "Jetway")

base.query('(Ruta == "IM-CUN")&(Proveedor == "Jetway")')

base.to_excel("/content/drive/MyDrive/Colab Notebooks/DataSets/Indice/Octubre/Xounting_Base_Oct_final.xlsx", sheet_name="Hoja1")

```
