## Prueba de Hipótesis para una Población

### Pruebas con Sigma Conocida

```python
import math
z = round((16/-15)/(2/math.sqrt(14)),2)
print("Z: ", z)
```

```python
from scipy.stats import norm
valor_p = 1- norm(0,1).cdf(z)
print("Valor p: ", valor_p)
```

### Pruebas con sigma desconocida

```python
import pandas as pd

datos = pd.Series([12,15,17,10,16,13,10,6,18,15,12,13,10,15])
media = datos.mean()
desv = round(datos.std(),2)
print("media: ",media)
print("Desviación: ", desv)
```

```python
import math
tn_1 = round((media-15)/(desv/math.sqrt(len(datos))),2)
print("Tn-1: ", tn_1)
```

```python
from scipy.stats import t
valor_p = t.cdf(tn_1, (len(datos)-1))
print("Valor p: ", valor_p)
```

### Pruebas para una Proporción

```python
import math
from scipy.stats import norm

x = 69
n = 153
p = round(x / n, 2)
print("p = ", p)

Z = round((p - 0.5) / math.sqrt(0.5 * (1 - 0.5) / n), 2)
print("Z:", Z)

valor_p = norm.cdf(Z)
print("valor-p: ", valor_p)
```
