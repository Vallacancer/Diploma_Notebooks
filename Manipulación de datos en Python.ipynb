### Concat y Append

```python
import pandas as pd

df1 = pd.DataFrame({"A":["A0", "A1", "A2", "A3"],
                    "B":["B0","B1","B2","B3"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]},
                   index=[0,1,2,3])

df4 = pd.DataFrame({"B":["B2","B3","B6","B7"],
                    "D":["D2","D3","D6","D7"],
                    "F":["F2","F3","F6","F7"]},
                   index=[2,3,6,7])

result = pd.concat([df1,df4], axis=1)
#result = result.set_axis(["A1","B1","C1","D1","B2","D2","F2"], axis = 1, copy=False)
print(result)
```

```python
import pandas as pd

df1 = pd.DataFrame({"A":["A0", "A1", "A2", "A3"],
                    "B":["B0","B1","B2","B3"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]},
                   index=[0,1,2,3])

df4 = pd.DataFrame({"B":["B2","B3","B6","B7"],
                    "D":["D2","D3","D6","D7"],
                    "F":["F2","F3","F6","F7"]},
                   index=[2,3,6,7])

result = pd.concat([df1,df4], axis=1, join="inner")
#result = result.set_axis(["A1","B1","C1","D1","B2","D2","F2"], axis = 1, copy=False)
print(result)#
```

Setting the index of columns

```python
import pandas as pd

df1 = pd.DataFrame({"A":["A0", "A1", "A2", "A3"],
                    "B":["B0","B1","B2","B3"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]},
                   index=[0,1,2,3])

df4 = pd.DataFrame({"B":["B2","B3","B6","B7"],
                    "D":["D2","D3","D6","D7"],
                    "F":["F2","F3","F6","F7"]},
                   index=[2,3,6,7])

result = pd.concat([df1,df4], axis=1)
result = result.set_axis(["A1","B1","C1","D1","B2","D2","F2"], axis = 1, copy=False)
print(result)
```

Right y Left join

```python
import pandas as pd

df1 = pd.DataFrame({"A":["A0", "A1", "A2", "A3"],
                    "B":["B0","B1","B2","B3"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]},
                   index=[0,1,2,3])

df4 = pd.DataFrame({"B":["B2","B3","B6","B7"],
                    "D":["D2","D3","D6","D7"],
                    "F":["F2","F3","F6","F7"]},
                   index=[2,3,6,7])

left_join = pd.concat([df1,df4], axis=1).reindex(df1.index)
right_join = pd.concat([df1,df4],axis=1).reindex(df4.index)
print(left_join, "\n\n", right_join)
```
