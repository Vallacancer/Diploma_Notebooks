### Regresión con Variables Transformadas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import seaborn as sns
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
x = np.linspace(0,5,50)
y = func(x, 2.5, 1.3, 0.5)
yn = y + 0.2*np.random.normal(size=len(x))
popt, pcov = curve_fit(func, x, yn)
plt.figure()
plt.plot(x, yn, 'ko', label="Datos originales")
plt.plot(x, func(x, *popt), 'r-', label="Curva de ajuste")
plt.legend()
plt.show()
```

## Distribuciones de Probabilidad

### Distribución de Probabilidad Uniforme

Seaborn

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np
# Crea un arreglo entre a= -4 hasta b= 8 en
# intervalor de 0.01.
a=-4
b=8
x = np.arange(a, b, 0.01)
y = uniform.pdf(x, 0, 5)
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel(r'$ \frac{1}{b-a} $')
plt.title('Distribución uniforme continua')
plt.show()
```

### Distribución Probabilidad continua

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
np.random.seed(2016)
sns.set_palette("deep",desat=0.6)
sns.set_context(rc={"figure.figsize":(8,4)})
# Graficando Normal
mu, sigma = 0, 1.6 # media y desvío estándar
mu1, sigma1 = 0, 0.8 # media y desvío estándar
mu2, sigma2 = 0, 0.4 # media y desvío estándar
normal = stats.norm(mu, sigma)
x = np.linspace(normal.ppf(0.01),
                normal.ppf(0.99), 100)
normal1 = stats.norm(mu1, sigma1)
x1 = np.linspace(normal1.ppf(0.01),
                normal1.ppf(0.99), 100)
normal2 = stats.norm(mu2, sigma2)
x2 = np.linspace(normal1.ppf(0.01),
                normal1.ppf(0.99), 100)
fp = normal.pdf(x) # Función de Probabilidad
fp1 = normal.pdf(x) # Función de Probabilidad
fp2 = normal.pdf(x2)
plt.plot(x,fp,x1,fp1,x2,fp2)
plt.title('Distribución Normal')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
