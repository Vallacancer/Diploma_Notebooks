```python
def horario_programado(base, puerto, serie, ruta, compañia="Ultramar"):
  for indice_base in base.query('(Embarcadero == @puerto) & (Ruta == @ruta) & (Proveedor == @compañia)')["Horario real"].index:
    chequeo = np.timedelta64(23,"h")
    for indice_serie in serie.index:
      if ruta == "Regresos":
        delta = abs(base["Horario real"][indice_base] - np.timedelta64(30, "m") - serie[indice_serie])
        base.loc[indice_base, "Ruta"] = "IM-CUN"
      else:
        delta = abs(base["Horario real"][indice_base] - serie[indice_serie])
        base.loc[indice_base, "Ruta"] = "CUN-IM"
      if delta < chequeo:
        chequeo = delta
        base.loc[indice_base, "Horario programado"] = serie[indice_serie]

# base["Horario real"] = pd.to_timedelta(base["Horario real"].astype(str))
# base["Horario programado"] = pd.to_timedelta(base["Horario programado"].astype(str))

#Se crean los horarios y se vuelven fechas de IM-CUN
playaTortugasIM = pd.to_datetime(pd.Series(["10:30:00", "11:45:00","13:00:00", "14:30:00", "16:00:00", "17:30:00","19:00:00",]))
playaCaracolIM = pd.to_datetime(pd.Series(["10:30:00","11:45:00","13:00:00","14:30:00","16:00:00","17:30:00","19:00:00"]))
PuertoJuarezUltramarIM = pd.to_datetime(pd.Series(["6:00:00","7:00:00","7:30:00","8:00:00","8:30:00","9:00:00","9:30:00","10:00:00",
                                  "10:30:00","11:00:00","11:30:00","12:00:00","12:30:00","13:00:00","13:30:00","14:00:00",
                                  "15:00:00","15:30:00","16:00:00","16:30:00","17:00:00","17:30:00","18:00:00","19:00:00",
                                  "19:30:00","20:00:00","21:00:00","22:00:00","23:00:00","00:00:00"]))
PuertoJuarezJetwayIM = pd.to_datetime(pd.Series(["7:00:00", "8:00:00", "9:00:00", "9:45:00", "10:30:00", "11:15:00", "12:00:00",
                                               "13:00:00", "14:00:00", "15:00:00", "16:00:00", "17:00:00", "18:00:00", "19:00:00",
                                               "20:00:00", "21:00:00"]))

#Se crean los horarios y se vuelven fechas de CUN-IM
playaTortugasCUN = pd.to_datetime(pd.Series(["9:30:00","11:00:00","12:15:00","13:30:00","15:00:00",
                                             "16:30:00","18:00:00"]))
playaCaracolCUN = pd.to_datetime(pd.Series(["9:45:00","11:15:00","12:30:00","13:45:00","15:15:00","16:45:00","18:15:00"]))
PuertoJuarezUltramarCUN = pd.to_datetime(pd.Series(["5:30:00","6:30:00","7:00:00","7:30:00","8:00:00","8:30:00","9:00:00","9:30:00",
                                                    "10:00:00","10:30:00","11:00:00","11:30:00","12:00:00","12:30:00","13:00:00","13:30:00",
                                                      "14:00:00","15:00:00","15:30:00","16:00:00","17:00:00","17:30:00","18:00:00","19:00:00","20:00:00",
                                                    "21:00:00","22:00:00","23:00:00"]))
PuertoJuarezJetwayCUN = pd.to_datetime(pd.Series(["6:00:00","7:00:00","8:00:00","9:00:00","9:45:00","10:30:00","11:15:00","12:00:00","13:00:00",
                                                  "14:00:00","15:00:00","16:00:00","17:00:00","18:00:00","19:00:00","20:00:00"]))

#Cambiar la serie a tiempo
#playaTortugas = pd.to_datetime(playaTortugas)

#Sumarle 1 hora
#playaTortugas = playaTortugas + np.timedelta64(1, "h")

#Accesar los datos de Playa Tortugas IM-CUN
#base.query('(Embarcadero == "Playa Tortugas") & (Ruta == "IM-CUN")')["Horario real"]

#Se hace la función para IM-CUN
horario_programado(base, "Playa Tortugas", playaTortugasIM, "Regresos")
horario_programado(base, "Playa Caracol", playaCaracolIM, "Regresos")
horario_programado(base, "Puerto Juarez", PuertoJuarezUltramarIM,"Regresos")
#Ya no existe Jetway
# horario_programado(base, "Puerto Juarez", PuertoJuarezJetwayIM,"Regresos", "Jetway")

#Se hace la función para CUN-IM
horario_programado(base, "Playa Tortugas", playaTortugasCUN, "Idas")
horario_programado(base, "Playa Caracol", playaCaracolCUN, "Idas")
horario_programado(base, "Puerto Juarez", PuertoJuarezUltramarCUN,"Idas")
# Ya no existe Jetway
# horario_programado(base, "Puerto Juarez", PuertoJuarezJetwayCUN,"Idas", "Jetway")
```

```python
from datetime import datetime
from datetime import date, timedelta

def horario_programado_COZ(base, serieA, serieB, ruta, compañia="Ultramar", puerto = "Playa del Carmen"):
  for indice_base in base.query('(Embarcadero == @puerto) & (Ruta == @ruta) & (Proveedor == @compañia)')["Horario real"].index:
    chequeo = np.timedelta64(23,"h")
    Ultramarprimeroref = date.fromisoformat("2023-11-02")
    dia = base.loc[indice_base,"Fecha muestra"]
    remainder = (Ultramarprimeroref- dia.date())% timedelta(days=2)
    if remainder == timedelta(days=1):
      for indice_serie in serieA.index:
        if ruta == "Regresos":
          delta = abs(base["Horario real"][indice_base] - np.timedelta64(50, "m") - serieA[indice_serie])
          base.loc[indice_base, "Ruta"] = "CUZ-PDC"
        else:
          delta = abs(base["Horario real"][indice_base] - serieA[indice_serie])
          base.loc[indice_base, "Ruta"] = "PDC-CUZ"
        if delta < chequeo:
          chequeo = delta
          base.loc[indice_base, "Horario programado"] = serieA[indice_serie]
    elif remainder == timedelta(days=0):
      for indice_serie in serieB.index:
        if ruta == "Regresos":
          delta = abs(base["Horario real"][indice_base] - np.timedelta64(50, "m") - serieB[indice_serie])
          base.loc[indice_base, "Ruta"] = "CUZ-PDC"
        else:
          delta = abs(base["Horario real"][indice_base] - serieB[indice_serie])
          base.loc[indice_base, "Ruta"] = "PDC-CUZ"
        if delta < chequeo:
          chequeo = delta
          base.loc[indice_base, "Horario programado"] = serieB[indice_serie]

#Horarios de Ultramar PDC-COZ
CozumelUltramar1PDC = pd.to_datetime(pd.Series(["8:00:00","10:00:00","12:00:00","14:00:00","16:00:00","18:00:00","20:00:00","22:00:00"]))
CozumelUltramar2PDC = pd.to_datetime(pd.Series(["9:00:00","11:00:00","13:00:00","15:00:00","17:00:00","19:00:00","21:00:00"]))

##Horarios de Ultramar COZ-PDC
CozumelUltramar1COZ = pd.to_datetime(pd.Series(["7:00:00","9:00:00","11:00:00","13:00:00","15:00:00","17:00:00","19:00:00","21:00:00"]))
CozumelUltramar2COZ = pd.to_datetime(pd.Series(["8:00:00","10:00:00","12:00:00","14:00:00","16:00:00","18:00:00","20:00:00"]))

#Horarios de Winjet PDC-COZ
CozumelWinjet2PDC = pd.to_datetime(pd.Series(["8:00:00","10:00:00","12:00:00","14:00:00","16:00:00","18:00:00","20:00:00","22:00:00"]))
CozumelWinjet1PDC = pd.to_datetime(pd.Series(["9:00:00","11:00:00","13:00:00","15:00:00","17:00:00","19:00:00","21:00:00"]))

##Horarios de Winjet COZ-PDC
CozumelWinjet2COZ = pd.to_datetime(pd.Series(["7:00:00","9:00:00","11:00:00","13:00:00","15:00:00","17:00:00","19:00:00","21:00:00"]))
CozumelWinjet1COZ = pd.to_datetime(pd.Series(["8:00:00","10:00:00","12:00:00","14:00:00","16:00:00","18:00:00","20:00:00"]))

horario_programado_COZ(base, CozumelUltramar1PDC, CozumelUltramar2PDC, "Idas")
horario_programado_COZ(base, CozumelUltramar1COZ, CozumelUltramar2COZ, "Regresos")

horario_programado_COZ(base, CozumelWinjet1PDC, CozumelWinjet2PDC, "Idas", "Winjet")
horario_programado_COZ(base, CozumelWinjet1COZ, CozumelWinjet2COZ, "Regresos", "Winjet")

base.loc[base["Embarcadero"]=="Playa del Carmen"].sample(10)
```
